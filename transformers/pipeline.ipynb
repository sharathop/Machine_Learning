```python

from transformers import pipeline

# 1. Sentiment Analysis
print("--- Sentiment ---")
print(pipeline("sentiment-analysis")("I love learning new things!"))

# 2. Text Generation
print("\n--- Text Generation ---")
print(pipeline("text-generation", model="gpt2")("Once upon a time", max_length=10))

# 3. Summarization
print("\n--- Summarization ---")
long_text = "Hugging Face is an AI community and platform. It provides tools that allow users to build, train and deploy machine learning models based on open source code and technologies."
print(pipeline("summarization")(long_text, max_length=15, min_length=5))

# 4. Question Answering
print("\n--- Question Answering ---")
qa = pipeline("question-answering")
print(qa(question="What is Hugging Face?", context=long_text))


#| Pipeline           | Underlying Model Class             |
# | ------------------ | ---------------------------------- |
# | sentiment-analysis | AutoModelForSequenceClassification |
# | text-generation    | AutoModelForCausalLM               |
# | summarization      | AutoModelForSeq2SeqLM              |
# | question-answering | AutoModelForQuestionAnswering      |
```
